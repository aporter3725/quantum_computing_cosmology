# Imports

```python
import numpy as np
from matplotlib import pyplot as plt
```

# Constants

```python
H = .5
T0 = 1.
TF = 100
RHO_T0 = 1./(8. * np.pi)
A0 = (3./2.)**(2./3.)
DOMAIN = np.arange(T0, TF + 1, H)
A = lambda t: ((3./2.) * t * np.sqrt(8. * np.pi * RHO_T0))**(2./3.)
A_PRIME = lambda a: np.sqrt((8. * np.pi * RHO_T0)/a)
```

# Functions

```python
def eulersMethod(a0, aPrime=A_PRIME, h=H):
    return a0 + h * aPrime(a0)
```

# Simulation

```python
solution = np.zeros_like(DOMAIN)
solution[0] = A0
for i in range(1, len(solution)):
    solution[i] = eulersMethod(solution[i-1])
plt.plot(DOMAIN, A(DOMAIN), 'k', linewidth=5, label='Solution')
plt.plot(DOMAIN, solution, 'c', label='Approximation')
plt.title('Euler\'s Method Approximation of Friedmann Equation')
plt.xlabel('$t$')
plt.ylabel('$a(t)$')
plt.legend(loc='best')
plt.show()
```
